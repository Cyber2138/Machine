```python
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = load_digits()
dir(data)
```

```python
plt.matshow(data.images[0])
plt.gray()
```

```python
categorical_features = data.data
target = data.target
```

```python
categorical_features.shape
```

```python
target.shape
```

Perfoming train and test spit function

```python
x_train, x_test, y_train, y_test = train_test_split(categorical_features, target, test_size=.2, random_state=0)
```

Building Decision tree classification model

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
dtree.score(x_test,y_test)
```

Decision tree model prediction

```python
y_pred = dtree.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot= True)
plt.title('Decision tree analysis')
plt.xlabel('Y_truth')
plt.ylabel('Y_predicted value')
plt.show()
```
