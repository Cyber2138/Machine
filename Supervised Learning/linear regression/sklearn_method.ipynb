```python
import pandas as pd 
```

```python
df = pd.read_csv('biryani.csv')
df
```

```python
from matplotlib import pyplot as plt 
x = df['Rice (KG)']
y = df['Persons']
plt.scatter(x,y)
plt.show()
```

```python
from scipy import stats

slope, intercepts, r,p, stand_err = stats.linregress(x,y)

def regression(n):
    return slope * n + intercepts

model = list(map(regression, x))

plt.scatter(x,y)
plt.plot(x, model)
plt.show()
```

```python
value  = regression(100)
value
```
