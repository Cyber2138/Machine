```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
logiregress  = np.poly1d(np.polyfit(x,y, deg=3))
line = np.linspace(0,22,num= 2000)

# plt.scatter(x,y)
plt.plot(line, logiregress(line))
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y, logiregress(x)))
```
