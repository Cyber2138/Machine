```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

data = load_iris()
print(data.__dir__())

print(data.target_names,'\n', data.target)
print()
print(data.feature_names,'\n', data.data)
```

```python
df = pd.DataFrame(data.data, columns= data.feature_names)
df
```

```python
df['target'] = data.target
df
```

```python
df['flowers'] = df['target']
```

```python
df['flowers'].replace([0,1,2], ['setosa', 'versicolor', 'virginica'], inplace=True)
df
```

```python
setosa = df[df.target ==0]
versicolor = df[df.target == 1]
virginica = df[df.target == 2]
```

```python
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], color= 'g', marker= 'x')
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], color= 'r', marker= 'x')
plt.xlabel('sepal lenght')
plt.ylabel('sepal width')
plt.show()
```

```python
plt.scatter(setosa['petal length (cm)'], setosa['petal width (cm)'], marker= 'x', color = 'r')
plt.scatter(versicolor['petal length (cm)'], versicolor['petal width (cm)'], marker= 'x', color = 'g')
```

```python
x = df.drop(['target', 'flowers'], axis= 1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= .2, random_state= 1)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()

log.fit(x_train, y_train)
```

```python
log.predict(x_test)
```

```python
log.score(x_test, y_test)
```
